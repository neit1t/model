```python
import pandas as pd

import sklearn as sk

from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import cross_val_score

from sklearn.pipeline import make_pipeline

from sklearn.svm import SVC

from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt

import numpy as np

import io

import requests
```

```python
url=["https://raw.githubusercontent.com/neit1t/model/main/titanic/titanic/train.csv",'https://raw.githubusercontent.com/neit1t/model/main/titanic/titanic/test.csv','https://raw.githubusercontent.com/neit1t/model/main/titanic/titanic/gender_submission.csv']

content_train,content_test_X,content_test_Y = requests.get(url[0]).content,requests.get(url[1]).content,requests.get(url[2]).content

train_dataset=pd.read_csv(io.StringIO(content_train.decode('utf-8')))
test_dataset = pd.read_csv(io.StringIO(content_test_X.decode('utf-8')))
y_test = pd.read_csv(io.StringIO(content_test_Y.decode('utf-8'))).iloc[:,1]
```

```python
corr = train_dataset.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
train_dataset['Age'].fillna(train_dataset['Age'].mean(),inplace=True)
train_dataset['Sex'].replace(['male','female'],[1,0],inplace = True)
test_dataset['Age'].fillna(test_dataset['Age'].mean(),inplace=True)
test_dataset['Sex'].replace(['male','female'],[1,0],inplace = True)
```

```python
train_dataset['Ravnopravie'] = train_dataset.apply(lambda x: 1 if (x['Sex'] == 1 and x['Age'] <= 16 ) or (x['Sex'] == 0) else 0,axis = 1)
train_dataset['Relatives'] = train_dataset.apply(lambda x: 0 if x['SibSp'] == 0 and x['Parch'] == 0 else 1,axis = 1)
test_dataset['Ravnopravie'] = test_dataset.apply(lambda x: 1 if (x['Sex'] == 1 and x['Age'] <= 16 ) or (x['Sex'] == 0) else 0,axis = 1)
test_dataset['Relatives'] = test_dataset.apply(lambda x: 0 if x['SibSp'] == 0 and x['Parch'] == 0 else 1,axis = 1)
```

```python
corr = train_dataset.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
train_dataset.drop(['PassengerId','Fare','Name','Ticket','Cabin','Embarked','SibSp','Parch'],axis = 1,inplace = True)
test_dataset.drop(['PassengerId','Fare','Name','Ticket','Cabin','Embarked','SibSp','Parch'],axis = 1,inplace = True)
```

```python
y_train = train_dataset.iloc[:,0]
X_train = train_dataset.iloc[:,1:]
X_test = test_dataset
```

```python
clf = make_pipeline(StandardScaler(), SVC(kernel = 'linear',C=0.0032,gamma = 'scale'))

clf.fit(X_train,y_train)


clf.score(X_test,y_test)
```

```python
sum(cross_val_score(clf,X_test,y_test, cv = 150))/150
```
